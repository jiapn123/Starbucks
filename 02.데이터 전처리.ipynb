```python
seoul_sgg = pd.read_excel("./seoul_sgg_list.xlsx")
seoul_sgg
```

```python
starbucks_list_byGu = pd.read_excel("./seoul_starbucks.xlsx",header=0)
print(starbucks_list_byGu.head())
```

```python
pivot_result =starbucks_list_byGu.pivot_table(index='시군구명', values='매장명', aggfunc='count')
pivot_result
```

```python
merged_df = pd.merge(seoul_sgg,pivot_result,how="left", on="시군구명")
merged_df
```

```python
sorted_df=merged_df.sort_values(by="매장명",ascending=False)
sorted_df
```

```python
seoul_sgg_pop = pd.read_excel("./sgg_pop.xlsx")
seoul_sgg_pop
```

```python
merged_df2 = pd.merge(merged_df,seoul_sgg_pop,how="left", on="시군구명")
merged_df2
```

```python
sgg_biz = pd.read_excel("./sgg_biz.xlsx")
sgg_biz
```

```python
merged_df3 = pd.merge(merged_df2,sgg_biz, how="left",on ="시군구명")
merged_df3 
```

```python
merged_df3.rename(columns={"매장명":"스탁벅스 매장수"},inplace=True)
merged_df3
```

```python
merged_df3.to_excel("./seoul_sgg_stat.xlsx",index=False)
```
